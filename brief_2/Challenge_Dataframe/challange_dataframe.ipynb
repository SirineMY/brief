```python
import pandas as pd 
#importer le fichier csv
df_google_ap_1 = pd.read_csv("googleplaystore_1.csv",sep=";")
df_google_ap_1
```

```python
df_google_ap_1[["Installs"]].value_counts()
```

```python

df_google_ap_1.Installs = df_google_ap_1.Installs.apply( lambda x: x.replace( "+", ""))
df_google_ap_1.Installs = df_google_ap_1.Installs.apply( lambda x: x.replace( "M", "000000"))
df_google_ap_1.Installs = df_google_ap_1.Installs.apply( lambda x: x.replace( ",", ""))

#filtrer la dataframe pour afficher les applications qui sont 1milliards+ d'installation
milliards=df_google_ap_1[df_google_ap_1["Installs"]=="1000000000"]
milliards[["App","Category","Genres","Installs"]]
```
