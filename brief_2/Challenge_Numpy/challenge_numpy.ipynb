```python
#importer les bibliothèque nécessaire

import pandas as pd 
import numpy as np
from scipy import linalg

#importer les fichiers 
df_marathon = pd.read_csv("marathon.txt", sep="\t", names = ["ville", "annee", "temps","secondes"])
df_marathon
```

```python
#ajouter les deux colonnees estPARIS et estBERLIN
df_marathon['estPARIS'] = df_marathon['ville'].apply(lambda x: 1 if x == "PARIS" else 0)
df_marathon['estBERLIN'] = df_marathon['ville'].apply(lambda x: 1 if x == "BERLIN" else 0)
df_marathon
```

```python
#création des matrice
X = df_marathon[["annee","estPARIS", "estBERLIN"]].to_numpy()
Y = df_marathon[["secondes"]].to_numpy()
```

```python
#calcule du coefficients
X_t = X.T
X_1 = np.dot(X_t,X)
X_1 = np.linalg.inv(X_1)
XX = np.dot(X_1,X_t)
A = np.matmul(XX,Y)
```

```python
#affichage des coefficients
print("a0=" + str(A[0]))
print("a1=" + str(A[1]))
print("a2=" + str(A[2]))
```
